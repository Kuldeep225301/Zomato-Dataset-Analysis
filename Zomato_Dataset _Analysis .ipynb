### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
plt.style.use('dark_background')
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df =df.drop(['url','address', 'phone', 'menu_item', 'reviews_list', 'dish_liked'], axis=1)
df.head()
```

```python
df.info()
```

### Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

###  Cleaning Rate columns

```python
df['rate'].unique()
```

### Removing "New", "-" and "/5" from Rate columns

```python
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()


# OR
# df['rate'] = df['rate'].apply(lambda x: np.nan if x in ['NEW', '-'] else float(str(x).split('/')[0]))
```

### Filling Null Values in rate Column with Mean

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace= True)
df['rate'].isnull().sum()
```

```python
round((df['rate'].isnull().sum() * 100) / len(df['rate']), 2)
```

```python
df.info()
```

### Droping Null Values

```python
df.dropna(inplace =True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Costplates','listed_in(type)':'Type'}, inplace= True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### Listed in(City) and location, both are there, lets keep only one

```python
df =df.drop(['listed_in(city)'], axis =1)
df.head()
```

```python
df['Costplates'].unique()
```

### Removing , from Costplates Columns

```python
def handlecomma(value):
    value= str(value)
    if',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

df['Costplates'] = df['Costplates'].apply(handlecomma)
df['Costplates'].unique()
# OR 
# df['Costplates'] = (df['Costplates'].astype(str).str.replace(',', '', regex=False).astype(float))
# df['Costplates'].unique()
```

```python
df.head()
```

### cleaning Rest type Column

```python
rest_type = df['rest_type'].value_counts()
rest_type
```

```python
rest_type = df['rest_type'].value_counts(ascending = False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type<1000]
rest_type_lessthan1000
```

### Making Rest types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] =  df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

### Cleaning Location Column

```python
location = df['location'].value_counts(ascending= False)
location
```

```python
location_lessthan300 = location[location <300]
location_lessthan300
```

```python
def handle_loaction(value):
    if(value in location_lessthan300):
        return 'Others'
    else:
        return value

df['location'] = df['location'].apply(handle_loaction)
df['location'].value_counts()
```

### Cleaning Cuisines Column

```python
df.columns
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisiness_lessthan100 = cuisines[cuisines<100]
cuisiness_lessthan100
```

```python
def handle_cuisiness(value):
    if(value in cuisiness_lessthan100):
        return 'others'
    else:
        return value

df['cuisines']= df['cuisines'].apply(handle_cuisiness)
df['cuisines'].value_counts()
```

```python
df.head()
```
